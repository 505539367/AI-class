```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
```

## **Part 1. Loading the dataset**

```python
# Using pandas load the dataset (load remotely, not locally)
dataset = pd.read_csv("https://raw.githubusercontent.com/profmcnich/example_notebook/main/science_data_large.csv")
# Output the first 15 rows of the data
dataset.head(15)
```

```python
# Display a summary of the table information (number of datapoints, etc.)
dataset.info()
dataset.describe()
```

```python
dataset.size #number of datapoint
```

## **Part 2. Splitting the dataset**

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = dataset[["Temperature °C", "Mols KCL"]]
X #Features
```

```python
y = dataset[["Size nm^3"]]
y #labels
```

```python
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

## **Part 3. Perform a Linear Regression**

```python
# Use sklearn to train a model on the training set

regr_model = LinearRegression()# Create linear regression object

regr_model.fit(X_train, y_train)# Train the model using the training sets

# Create a sample datapoint and predict the output of that sample with the trained model

regr_model.predict(X_test) # Make predictions
```

```python
# Report on the score for that model, in your own words (markdown, not code) explain what the score means
score = regr_model.score(X_train, y_train)
print("Score for model :", score)
print("Standard devisation :", score.std())
```

### Score:
The score of the model is between 0 to 1 (0% to 100%). So if it is 100%, the two variables are perfectly correlated. If it is 0%, the two variables are not correlated at all. The score for the model above is around 0.86, two variable has 86% correlation.

```python
# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
print("Coeffients :", regr_model.coef_) 
print("Intercept :", regr_model.intercept_) 
```

### Equation for h(x):
$h(x) = 884.88696094x_0 + 1022.31595498x_1 - 417120.29306632$

where $x_0$ is Temperature °C and $x_1$ is Mols KCL.

## **Part 4. Use Cross Validation**

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# Report on their finding and their significance

print("Use the cross_val_score with 3 folds:")

scores = cross_val_score(regr_model ,X_train, y_train, cv = 3)
print("\tCross_val_scores :",scores)
print("\tAccracy :",scores.mean())
print("\tStandard devisation :", scores.std())
```

```python
print("Use the cross_val_score with 5 folds:")

scores = cross_val_score(regr_model ,X_train, y_train, cv = 5)
print("\tCross_val_scores :",scores)
print("\tAccracy :",scores.mean())
print("\tStandard devisation :", scores.std())
```

```python
print("Use the cross_val_score with 7 folds:")

scores = cross_val_score(regr_model ,X_train, y_train, cv = 7)
print("\tCross_val_scores :",scores)
print("\tAccracy :",scores.mean())
print("\tStandard devisation :", scores.std())
```

### Cross Validation finding:
Two variable always has around 86% correlation, and standard devisation is increasing as I tried to split the data into more partes.

## **Part 5. Using Polynomial Regression**

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(2)
feature = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(feature, y, test_size = 0.1) # split the feature and label
poly_model = regr_model
poly_model.fit(X_train, y_train)

# Make predictions
poly_model.predict(X_test) 
```

```python
# Report on the metrics and output the resultant equation as you did in Part 3.
score = poly_model.score(X_train, y_train)
print("Score for model :", score)
print("Coeffients :", poly_model.coef_) 
print("Intercept :", poly_model.intercept_) 
```

### Equation for h(x):

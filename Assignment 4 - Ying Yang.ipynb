```python
# Imports and pip installations (if needed)
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings("ignore")
```

## Part 1: Load the dataset

```python
# Load the dataset (load remotely, not locally)
dataset = load_iris()
dataframe = pd.DataFrame(data = dataset['data'],
                     columns= dataset['feature_names'])
dataframe["target"] = dataset.target
# Output the first 15 rows of the data
dataframe.head(15)
```

```python
# Display a summary of the table information (number of datapoints, etc.)
print(dataset.target_names)
```

```python
print(dataset.feature_names)
```

```python
dataframe.info()
```

```python
dataframe.describe()
```

### About the dataset
There are 150 samples from the iris dataset.
Included three species of Iris:

    0 represent Iris-Setosa
    1 represent Iris-Versicolour
    2 represent Iris-Virginica

The features should be dataset['data'], which contains sepal length, sepal width, petal length, petal width.

The label will be target, which has value of 0, 1, and 2. The value is representing a species of Iris.




## Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = dataframe[dataset.feature_names]
X #Features
```

```python
y = dataframe["target"]
y #labels
```

```python
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
X_train
```

```python
X_test
```

## Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
LR_model = LogisticRegression()
LR_model.fit(X_train, y_train)
```

```python
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_point = X[:1]
sample_pred = LR_model.predict_proba(sample_point)[0]

print("Class\t\tSpecies\t\t\tProbability")
print("0\t\tSetosa\t\t\t{:0.5f}".format(sample_pred[0]))
print("1\t\tVersicolor\t\t{:0.5f}".format(sample_pred[1]))
print("2\t\tVirginica\t\t{:0.5f}".format(sample_pred[2]))
```

```python
#Predict the x_test data set, and get the probabilities for each possible class
print("y_test\t\tProbability of Setosa\tProbability of Versicolor\tProbability of Virginica",end='')
index = 0;
for i in LR_model.predict(X_test):
    print("\n",i,end ='')
    for j in LR_model.predict_proba(X_test)[index]:
        print("\t\t\t{:0.5f}".format(j),end ='')
    index+= 1;
```

All the prediction are correct in the result above.

```python
# iii. Report on the score for Logistic regression model, what does the score measure?
train_score = LR_model.score(X_train, y_train)
test_score  = LR_model.score(X_test, y_test)
print("Training score for model :", train_score)
print("Test score for model :", test_score)
```

### Score:
The score of the model is between 0 to 1 (0% to 100%) and it represent the accuracy on the given data. 
the score of training data is 0.97, means the accuracy on the model is about 97%.
The score of test data is 1.0, means model is perfectly accurate.

```python
# iv. Extract the coefficents and intercepts for the boundary line(s)
print("Coeffients :", LR_model.coef_) 
print("Intercept :", LR_model.intercept_) 
```

All the predictions are correct in the result above

```python
# iii. Report on the score for the SVM, what does the score measure?
train_score = SVM_model.score(X_train, y_train)
test_score  = SVM_model.score(X_test, y_test)
print("Training score for model :", train_score)
print("Test score for model :", test_score)
```

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    
    while (low <= high):
 
        mid = (high + low) // 2
 
        # When x is grater than 'mid'
        if (target > arr[mid]):
            low = mid + 1
 
        # When x is grater than 'mid'
        elif (target < arr[mid]):
            high = mid - 1
 
        # when target is found
        else:
            return mid
 
    # When target no exist in arr
    return -1
 
```

```python
#Test
arr = [ 0, 1, 2, 3, 4, 8, 10, 15, 40, 75, 100 ]
target = 15
 
result = binary_search(arr, target)
 
if result == -1:
    print("Element is not exist in array")
else:
    print("Element found! At index:", result);
```

## **Binary Search:**

The input of the function are: arr, target.
* arr is an array with n elemenyts.
* target is the element we trying to find its index in the arr

The function will return index of target if it exist in arr, otherwise will return 0.

An array will have n indexes, which is equal to the length of the arr - 1.

Then we initialize two variables, which are 'low' and 'high'. 'low' correspond to the first index in the arr that we going to search, the 'high' correspond to the last index in the arr.

Create a while loop, when the loop is continue, the program will chek if the target is equal to the middle index between index low and high.

There are three cases exist while checking the indexes:
* When the target is greater than the middle index between 'low' and 'high', then the target might exist after the middle index, thus make the current middle index as the new 'low' value.
* When the target is less than the middle index between 'low' and 'high', then the target might exist before the middle index, thus make the current middle index as the new 'high' value.
* When the target is equal to the middle index between 'low' and 'high', it means the target is found, the return the middle index.

The while loop stop when there is no remain index between 'low' and 'high', means the program already checked all possible indexes and the target is not found. Then return -1.
